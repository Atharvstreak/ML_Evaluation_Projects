# Problem Statement

Coronavirus is a family of viruses that can cause illness, which can vary from common cold and cough to sometimes more severe disease. Middle East Respiratory Syndrome (MERS-CoV) and Severe Acute Respiratory Syndrome (SARS-CoV) were such severe cases with the world already has faced. SARS-CoV-2 (n-coronavirus) is the new virus of the coronavirus family, which first discovered in 2019, which has not been identified in humans before. It is a contiguous virus which started from Wuhan in December 2019. Which later declared as Pandemic by WHO due to high rate spreads throughout the world. Currently (on the date 20 May 2020), this leads to a total of 300K+ Deaths across the globe, including 90K+ deaths alone in USA.

This is the data for the 2019 Novel Coronavirus Visual Dashboard given by the Johns Hopkins University Center for Systems Science and Engineering (JHU CSSE). Also, Supported by ESRI Living Atlas Team and the Johns Hopkins University Applied Physics Lab (JHU APL).

In this project we will predict the Hospitalization Rate.

# Import needed modules and get the required dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import glob #the glob module is used to retrieve files/pathnames matching a specified pattern
#path for files
path = r'C:\Users\admin\Downloads\COVID-19-master\csse_covid_19_data\csse_covid_19_daily_reports_us'
all_files = glob.glob(path + '\*.csv')
#loop to create a list of all dataframes containing data from all the csv
dfs = []
for f in all_files:
    df = pd.read_csv(f)
    dfs.append(df)
#concatenate all the dfs
covid19_us_dat = pd.concat(dfs, ignore_index=True)
```

```python
covid19_us_dat #check data
```

Feature Descirption :

1. Province_State - The name of the State within the USA.
2. Country_Region - The name of the Country (US).
3. Last_Update - The most recent date the file was pushed.
4. Lat - Latitude.
5. Long_ - Longitude.
6. Confirmed - Aggregated confirmed case count for the state.
7. Deaths - Aggregated Death case count for the state.
8. Recovered - Aggregated Recovered case count for the state.
9. Active - Aggregated confirmed cases that have not been resolved (Active = Confirmed - Recovered - Deaths).
10. FIPS - Federal Information Processing Standards code that uniquely identifies counties within the USA.
11. Incident_Rate - confirmed cases per 100,000 persons.
12. People_Tested - Total number of people who have been tested.
13. People_Hospitalized - Total number of people hospitalized.
14. Mortality_Rate - Number recorded deaths * 100/ Number confirmed cases.
15. UID - Unique Identifier for each row entry.
16. ISO3 - Officialy assigned country code identifiers.
17. Testing_Rate - Total number of people tested per 100,000 persons.

Target variable :

Hospitalization_Rate - Total number of people hospitalized * 100/ Number of confirmed cases.

A lot of columns have significantly large number of missinf values!

Unfortunately we have 10,000+ values out of 15,000+ values missing in our target variable... this means 10,000+ rows having these missing values won't help our machine to learn. Sadly we'll have to discard this much data :(

```python
#removing all the rows where target variable is missing
covid19_us_dat.dropna(how='any', subset=['Hospitalization_Rate'],inplace=True)
```

```python
#check null values after above changes
covid19_us_dat.info()
```

```python
#check shape after above changes
covid19_us_dat.shape
```

We can clearly see, most of the columns which were having missing values earlier have no missing values now.

'Recovered' column still has a significant amount of missing values , we should impute these values.

```python
#imputing missing values in 'Recovered'
covid19_us_dat['Recovered'].fillna(covid19_us_dat['Recovered'].mean(),inplace=True)
```

```python
#imputing missing value in 'FIPS'
covid19_us_dat['FIPS'].fillna(covid19_us_dat['FIPS'].mean(),inplace=True)
```

```python
#imputing missing value in 'Last_Update'
covid19_us_dat['Last_Update'].fillna(covid19_us_dat['Last_Update'].mode()[0],inplace=True)
```

```python
#'Last_Update' is of dtype 'object' , we know that this feature gives information about date and time 
# convert 'Last_Update' to dtype 'datetime'
covid19_us_dat['Last_Update']=pd.to_datetime(covid19_us_dat['Last_Update'])
```

```python
# According to our problem statement , only the month and week of the year seems important for training our model 
# let's create two new columns as 'month' and 'week of year' from 'Last_update' column and later delete 'Last_Update'
covid19_us_dat['Month']=covid19_us_dat.Last_Update.dt.month
covid19_us_dat['WeekOfYear']=covid19_us_dat.Last_Update.dt.weekofyear
covid19_us_dat.drop('Last_Update',axis=1,inplace=True)
```

# Preparing data for training

```python
#removing redundant coloumns
covid19_us_dat.drop(['Province_State','Country_Region','FIPS','UID','ISO3'],axis=1,inplace=True)
```

```python
#importing zscore to treat outliers
import scipy
from scipy.stats import zscore
z = abs(zscore(covid19_us_dat))
covid19_us_dat[z>3] #detecting outliers
```

We need to remove all these outliers, because regression models are sensitive to outliers.

```python
#remove outliers
covid19_us_dat = covid19_us_dat[(z<3).all(axis=1)]
```

```python
#check skewness of data
covid19_us_dat.hist(grid=True,bins=20,layout=(7,5),figsize=(20,20))
plt.show()
covid19_us_dat.skew()
```

The skewness of most the variables need to be improved. 

The target variable is also slightly skewed.

Let us remove skewness where columns have skewness corfficient of more than 0.5

```python
#the data has negative values, we must convert them to positive values before applying log transformation
covid19_us_dat = covid19_us_dat.abs()
```

```python
#seperating target variable
x = covid19_us_dat.drop('Hospitalization_Rate',axis=1)
y = covid19_us_dat['Hospitalization_Rate']
```

```python
#reduce skewness of columns with skewness coeff more than 0.5 using log
for i in x.columns:
    if x[i].dtype in ['int64','float64']:
        if x[i].skew() > 0.5:
            x[i]=np.log1p(x[i])
        if x[i].skew() < -0.5:
            x[i]=np.square(x[i])
x.skew()#check improved skewness
```

```python
#plotting correlation heatmap
plt.figure(figsize=(12,6))
sns.heatmap(covid19_us_dat.corr(),annot=True)
plt.show()
```

We can see that confirmed, deaths, recovered, active, incident_rate, people_tested, people_hospitalized are highly correlated with each other, and testing_rate, month and year are also highly correlated to each other ,to avoid the problem of mulitcolinearity, we should apply PCA on such features.

'mortality rate' has the highest correlation with the target variable among all features.

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head()
```

```python
#Principal Component Analysis
from sklearn.decomposition import PCA
#apply PCA to confirmed, deaths, recovered, active, incident_rate, people_tested, people_hospitalized
pca_1 = PCA(n_components=4)
temp_1 = pd.DataFrame(pca_1.fit_transform(x[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'People_Tested',
                                           'People_Hospitalized']]),columns=['pca_1','pca_2','pca_3','pca_4'])
#apply PCA to testing_rate, month and week of year
pca_2 = PCA(n_components=2)
temp_2 = pd.DataFrame(pca_2.fit_transform(x[['Testing_Rate', 'Month', 'WeekOfYear']]),columns=['pca_5','pca_5'])
#combing all the pca components into data and dropping all the features used to create these components 
x = pd.concat((x, temp_1, temp_2), axis=1)
x.drop(['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'People_Tested','People_Hospitalized','Testing_Rate',
        'Month', 'WeekOfYear'], axis=1, inplace=True)
x.head() #check data
```

```python
#train-test-split
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#importing all the regression algorithms
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#importing all the metrics and tuning methods for regression analysis
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=None)
#selecting the algorithm which works best 
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor()]
for m in models:
    m.fit(x_train,y_train)
    print(m)
    print('SCORE :',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('r2_score : ',r2_score(y_test,pred))
    print('mean absolute error :',mean_absolute_error(y_test,pred))
    print('mean squared error :',mean_squared_error(y_test,pred))
    print('root mean squared error :',np.sqrt(mean_squared_error(y_test,pred)))
    print('\n')
```

KNeighborsRegressor and DecisionTreeRegressor are giving best results.

1) KNeighborsRegressor

```python
#checking the accuracy score of KNeighborsRegressor at best random_state
knn=KNeighborsRegressor()
max_score_knn=0
best_ran_state_knn=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    knn.fit(x_train,y_train)
    pred=knn.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_knn < score:
        max_score_knn = score
        best_ran_state_knn = n
print('max r2_score of KNeighborsRegressor : ',max_score_knn)
print('best random state : ',best_ran_state_knn)
```

2) DecisionTreeRegressor

```python
#checking the accuracy score of DecisionTreeRegressor at best random_state
dt=DecisionTreeRegressor()
max_score_dt=0
best_ran_state_dt=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    dt.fit(x_train,y_train)
    pred=dt.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_dt < score:
        max_score_dt = score
        best_ran_state_dt = n
print('max r2_score of DecisionTreeRegressor : ',max_score_dt)
print('best random state : ',best_ran_state_dt)
```

Clearly, KNeighbors is working slightly better than DecisionTree, at random state 58.

So we will finally select KNRegressor for final training.

```python
#tuning the hyperparameters of KNN with best random state
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=58)
params = [{'n_neighbors': [2,3,4,5,6], 'weights': ['uniform','distance']}]
KNN=KNeighborsRegressor()
GCV=GridSearchCV(estimator=KNN,param_grid=params,cv=10)
GCV.fit(x_train,y_train)
```

```python
#getting best parameters
print('best parameters are : ',GCV.best_params_)
```

```python
#final training and testing with best parameters
KNN_best=KNeighborsRegressor(n_neighbors=3,weights='distance')
KNN_best.fit(x_train,y_train)
preds=KNN_best.predict(x_test)
print('SCORE :',KNN_best.score(x_train,y_train))
print('r2_score : ',r2_score(y_test,preds))
print('mean absolute error :',mean_absolute_error(y_test,preds))
print('mean squared error :',mean_squared_error(y_test,preds))
print('root mean squared error :',np.sqrt(mean_squared_error(y_test,preds)))
```

We got a tuned KNN algroithm now.The metrics returned improved values now!

```python
#check whether model is overfitting or not
scores=cross_val_score(KNN_best,x_train,y_train,cv=10,scoring='r2')
print('mean r2 score : ',scores.mean())
```

Model is not overfitting

```python
#plotting predicted v/s actual values
sns.distplot(preds,label='predicted values')
sns.distplot(y_test,label='actual values')
plt.title('predicted v/s actual values')
plt.legend()
plt.show()
```

```python
#checking important features
from sklearn.inspection import permutation_importance
plt.figure(figsize=(12,6))
imps = permutation_importance(KNN_best,x,y)
imp = pd.Series(imps.importances_mean, index=x.columns)
imp.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```

'pca_1' is the most important feature, pca_5 is the least important.
