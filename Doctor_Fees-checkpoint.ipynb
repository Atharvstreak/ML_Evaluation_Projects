# Problem Statement

We have all been in situation where we go to a doctor in emergency and find that the consultation fees are too high. As a data scientist we all should do better. What if you have data that records important details about a doctor and you get to build a model to predict the doctor’s consulting fee.?

# Import needed modules and get the required dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#get the data
doc_dat = pd.read_excel('Downloads/Final Participant Data Folder/Final_Train.xlsx')
doc_dat.to_csv('Downloads/Final Participant Data Folder/Final_Train.csv')
doc_dat.head() #check data
```

FEATURE Description:

Qualification: Qualification and degrees held by the doctor

Experience: Experience of the doctor in number of years

Rating: Rating given by patients

Profile: Type of the doctor

Miscellaeous_Info: Extra information about the doctor

Place: Area and the city where the doctor is located.

Target Variable :

Fees: Fees charged by the doctor

```python
#let's check null values in 'Qualification' column
print('null values : ',doc_dat['Qualification'].isnull().sum())
```

no missing values present.

```python
#check unique values in qualification
for i in range(len(doc_dat['Qualification'].unique())):
    print(doc_dat['Qualification'].unique()[i])
```

We can see that the qualifications are entered in a jumbled way, let us create new columns of most common qualifications.

```python
#Let us now convert this single column('Qualification') into multiple columns 
#where each column will have one single(most common) education level
most_common = ['BSc','BDS','BHMS','BAMS','MSc','MDS','MBBS','MD','MS','DNB','PhD','DDVL','DDV','DVD',
               'Diploma','Fellow','Certifi','PostGraduate','PG','Master','Venereology&Leprosy']
for i in range(len(most_common)) :
    temp = most_common[i]
    temp_0 = []
    for j in range(len(doc_dat['Qualification'])):
        doc_dat['Qualification'][j] = ((doc_dat['Qualification'][j].replace('.','')).replace(' ','')).replace('-','')
        temp_1 = doc_dat['Qualification'][j].split(',')
        temp_2 = 0
        for k in temp_1:
            if (k.lower()).__contains__(temp.lower()):
                temp_2 = temp_2 + 1
                doc_dat['Qualification'][j] = doc_dat['Qualification'][j].replace(k,'')
        if temp_2 > 0:
            temp_0.append(1)
        else:
            temp_0.append(0)
    doc_dat[temp] = temp_0
#remove all commas
for r in range(len(doc_dat['Qualification'])):
    doc_dat['Qualification'][r] = doc_dat['Qualification'][r].replace(',','')
```

```python
doc_dat.iloc[:,7:].head() #check_data
```

```python
#add similar columns
doc_dat['Diploma']=doc_dat['DNB']+doc_dat['DDVL']+doc_dat['DDV']+doc_dat['DVD']+doc_dat['Diploma']+doc_dat['Venereology&Leprosy']
doc_dat['Masters']=doc_dat['Master']+doc_dat['PostGraduate']+doc_dat['PG']
doc_dat.drop(['DNB','DDVL','DDV','DVD','PG',
              'PostGraduate','Master','Venereology&Leprosy'],axis=1,inplace=True)
#check changes
doc_dat.head()
```

```python
#seperate target variable 
x = doc_dat.drop('Fees',axis=1)
y = doc_dat['Fees']
```

```python
#label encode places data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for c in x.columns:
    if c in ['Place_0','Place_1']:
        x[c] = le.fit_transform(x[c])
#one hot encode 
x = pd.get_dummies(x)
x.head()
```

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head()
```

```python
#train-test-split
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#importing all the regression algorithms
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#importing all the metrics and tuning methods for regression analysis
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.compose import TransformedTargetRegressor
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=None)
#selecting the algorithm which works best 
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor(),
       RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),XGBRegressor(verbosity=0)]
for m in models:
    m.fit(x_train,y_train)
    print(m)
    print('SCORE :',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('r2_score : ',r2_score(y_test,pred))
    print('mean absolute error :',mean_absolute_error(y_test,pred))
    print('mean squared error :',mean_squared_error(y_test,pred))
    print('root mean squared error :',np.sqrt(mean_squared_error(y_test,pred)))
    print('\n')
```

After analysing above results, we can say that GradientBoostingRegressor is giving less errors than others.

So finally, let us select GBRegressor for final training.

```python
#checking the accuracy score of GBRegressor at best random_state
gbr=GradientBoostingRegressor()
max_score_gbr=0
best_ran_state_gbr=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    gbr.fit(x_train,y_train)
    pred=gbr.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_gbr < score:
        max_score_gbr = score
        best_ran_state_gbr = n
print('max r2_score of GBRegressor : ',max_score_gbr)
print('best random state : ',best_ran_state_gbr)
```

```python
#let us try LightGBRegressor
import lightgbm 
from lightgbm import LGBMRegressor
lgbr = LGBMRegressor(silent=True)
max_score_lgbr=0
best_ran_state_lgbr=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    lgbr.fit(x_train,y_train)
    pred=lgbr.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_lgbr < score:
        max_score_lgbr = score
        best_ran_state_lgbr = n
print('max r2_score of LGBRegressor : ',max_score_lgbr)
print('best random state : ',best_ran_state_lgbr)
```

LGBM is performing better than GBRegressor , let us select LGBMRegressor for final training.

```python
#let's tune parameters of lgbr
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=81)
param = {
        "max_depth": [25,50, 75],"learning_rate" : [0.01,0.05,0.1],"num_leaves": [300,900,1200],"n_estimators": [200]
        }
GCV_lgbr= GridSearchCV(lgbr,param_grid=param,cv=10,scoring='r2')
GCV_lgbr.fit(x_train,y_train)
```

```python
#get best params for lgbr
GCV_lgbr.best_params_
```

```python
#final training and testing with best parameters
lgbr_best = LGBMRegressor(learning_rate= 0.01,max_depth=25,n_estimators=200,num_leaves=300,silent=True)
lgbr_best.fit(x_train,y_train)
preds=lgbr_best.predict(x_test)
print('SCORE :',lgbr_best.score(x_train,y_train))
print('r2_score : ',r2_score(y_test,preds))
print('mean absolute error :',mean_absolute_error(y_test,preds))
print('mean squared error :',mean_squared_error(y_test,preds))
print('root mean squared error :',np.sqrt(mean_squared_error(y_test,preds)))
```

```python
#cross_validation
scores=cross_val_score(lgbr_best,x,y,cv=10,scoring='r2')
print('mean r2 score : ',scores.mean())
```

```python
#plotting predicted v/s actual values
sns.distplot(preds,label='predicted values')
sns.distplot(y_test,label='actual values')
plt.title('predicted v/s actual values')
plt.legend()
plt.show()
```

We can see that our model is not fitting properly and is predicting less fees when the actual fees is above 300 and predicting greater fees when actual fees is below 300.

```python
#finally fiiting 100% data for deployment
lgbr_best.fit(x,y)
lgbr_best.score(x,y)
```

```python
#saving the model using joblib
import joblib 
joblib.dump(lgbr_best,'Doctor_Fees.pkl')
```
