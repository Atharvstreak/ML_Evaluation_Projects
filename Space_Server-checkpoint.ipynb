# Import needed modules and get the required dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# get the dataset
space_dat = pd.read_csv('space_server.csv')
space_dat.head()
```

Feature Descirption :

1. objid = Object Identifier
2. ra = J2000 Right Ascension (r-band) 
        (Right ascension (abbreviated RA) is the angular distance measured eastward along the celestial equator from the     Sun at  the March equinox to the hour circle of the point above the earth in question.) 
3. dec = J2000 Declination (r-band)
        (When 'ra' paired with declination (abbreviated dec), these astronomical coordinates specify the direction of a    point on the celestial sphere (traditionally called in English the skies or the sky) in the equatorial coordinate       system.)
4. u, g, r, i, z :
       The Thuan-Gunn astronomic magnitude system. u, g, r, i, z represent the response of the 5 bands of the telescope. 
5. Run, rerun, camcol and field :
       Describes a field within an image taken by the SDSS. A field is basically a part of the entire image corresponding  to  2048 by 1489 pixels.
       A field can be identified by:
               - run number, which identifies the specific scan,
               - the camera column, or "camcol," a number from 1 to 6, identifying the scanline within the run, and the    field number. 
               - The field number typically starts at 11 (after an initial rampup time), and can be as large as 800 for      particularly long runs.
               - rerun, specifies how the image was processed.
6. specobjid = Object Identifier
7. redshift = Final Redshift 
        (redshift happens when light or other electromagnetic radiation from an object is increased in wavelength, or       shifted to the red end of the spectrum.)
8. plate = plate number 
        (Each spectroscopic exposure employs a large, thin, circular metal plate that positions optical fibers via holes   drilled at the locations of the images in the telescope focal plane. These fibers then feed into the spectrographs. Each plate has a unique serial number, which is called plate in views such as SpecObj in the CAS.)
9. mjd = MJD of observation
        (Modified Julian Date, used to indicate the date that a given piece of SDSS data (image or spectrum) was taken.)
10. fiberid = fiber ID
        (The SDSS spectrograph uses optical fibers to direct the light at the focal plane from individual objects to the    slithead. Each object is assigned a corresponding fiberID.)
        
TARGET VARIABLE :

class = object class (galaxy, star or quasar object)

```python
#removing redundant columns
space_dat.drop(['objid','rerun','specobjid'],axis=1,inplace=True)
space_dat.shape
```

```python
#encode target varible (class)
#import label encoder from sklearn
import sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
space_dat['class']=le.fit_transform(space_dat['class'])
```

```python
#check shape
space_dat.shape
```

```python
#importing zscore to treat outliers
import scipy
from scipy.stats import zscore
z = abs(zscore(space_dat))
space_dat[z>3] #detecting outliers
```

We can see outliers present in this data, but they are significantly very large in number, so we will let them be and not remove them. Also the algorithms that we are going to use are not sensitive to outliers.

```python
#plotting correlation heatmap
plt.figure(figsize=(20,8))
sns.heatmap(space_dat.corr(),annot=True)
plt.show()
```

We can see that 'mjd' and 'plate' are highly correlated with each other, just like u,g,r,i,z are highly correlated with each other, to avoid the problem of mulitcolinearity, we should apply PCA on such features. 

'mjd' has the highest correlation with the target variable among all features.

```python
#seperating target variable
x = space_dat.drop('class',axis=1)
y = space_dat['class']
```

```python
#check skewness
x.skew()
```

Some data is skewed.

'redshift' is highly skewed.

Since this is a classification problem , it is not necessary to remove skewness, as the tree based models do not assume the variables to be normally distributed.

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head()
```

```python
#Principal Component Analysis
from sklearn.decomposition import PCA
#apply PCA to u,g,r,i,z
pca = PCA(n_components=3)
astro_mags = pd.DataFrame(pca.fit_transform(x[['u', 'g', 'r', 'i', 'z']]),columns=['pca_1','pca_2','pca_3'])
#apply PCA to plate and mjd
pca_1 = PCA(n_components=1)
plate_mjd = pd.DataFrame(pca_1.fit_transform(x[['plate','mjd']]),columns=['pca_4'])
#combing all the pca components into space_dat and dropping all the features used to create these components 
x = pd.concat((x, astro_mags, plate_mjd), axis=1)
x.drop(['u', 'g', 'r', 'i', 'z','plate','mjd'], axis=1, inplace=True)
x.head() #check data
```

```python
#importing all the classification algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
#importing all the metrics for classification analysis
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score
```

```python
#import train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=None)
#import cross val score 
from sklearn.model_selection import cross_val_score
model_list=[KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),XGBClassifier(verbosity=0)]
for model in model_list:
    results=cross_val_score(model,x_train,y_train,cv=10,scoring='accuracy')
    print(model)
    print('mean score : ',results.mean())
    print('\n')
```

Clearly, all the algorithms are working great. XGBClassifier is working best among all.

```python
#checking the accuracy score of XGBClassifier at best random_state
xgbc=XGBClassifier(verbosity=0)
max_score_xgbc=0
best_ran_state_xgbc=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=n)
    xgbc.fit(x_train,y_train)
    pred=xgbc.predict(x_test)
    score=accuracy_score(y_test,pred)
    if max_score_xgbc < score:
        max_score_xgbc = score
        best_ran_state_xgbc = n
print('max accuracy score of XGBClassifier : ',max_score_xgbc)
print('best random state : ',best_ran_state_xgbc)
```

We must now find the best parameters for XGBClassifier, so that we can be sure that our model is not overfiting.

```python
#train test split with best random state obtained for XGBClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=62)
#importing GridSearchCV 
from sklearn.model_selection import GridSearchCV
#tuning hyperparameters of XGBClassifier
hyper_params={"max_depth": [ 3, 4, 5, 6, 8, 10, 12, 15],
              "min_child_weight" : [ 1, 3, 5, 7 ],
              "gamma": [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
              "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ] }
GCV_XGBC=GridSearchCV(estimator=xgbc,param_grid=hyper_params,cv=10)
GCV_XGBC.fit(x_train,y_train)
```

```python
GCV_XGBC.best_params_
```

```python
#training and testing with best parameters of KNeighborsClassifier
xgbc_best=XGBClassifier(colsample_bytree=0.7,gamma=0,max_depth=3,min_child_weight=1)
xgbc_best.fit(x_train,y_train)
predictions=xgbc_best.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,predictions))
print('confusion matrix : ','\n',confusion_matrix(y_test,predictions))
print('classification report : ','\n',classification_report(y_test,predictions))
```

```python
#plotting roc_curve 
pred_prob=xgbc_best.predict_proba(x_test)[:,1]
fpr,tpr,threshholds=roc_curve(y_test,pred_prob,pos_label=True)
plt.plot(fpr,tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc_curve')
plt.show()
```

```python
#let's see the most important features
(pd.Series(xgbc_best.feature_importances_,index=x.columns)).plot(kind='bar')
plt.title('Feature importances')
plt.show()
```

'redshift' is the most important feature followed by 'pca_4', then comes 'pca_3','pca_2','pca_1' respectively.

It seems like 'ra','dec','run','camcol','field' and 'fibrid' are not that significantly important.

'camcol' shows 0 importance.

```python
#finally fitting 100% data for deployment
xgbc_best.fit(x,y)
xgbc_best.score(x,y)
```

```python
#saving the model using joblib
import joblib 
joblib.dump(xgbc_best,'space_server.pkl')
```
