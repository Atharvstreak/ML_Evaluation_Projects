```python
#remove outliers
train=train[(z<3).all(axis=1)]
```

```python
#check skewness of data
train[['VOTES','COST']].hist(grid=True,bins=20,layout=(1,2),figsize=(5,5))
plt.show()
train[['VOTES','COST']].skew()
```

```python
#reduce skewness 
for s in ['VOTES','COST']:
    if train[s].skew() > 0.5:
        train[s]=np.log1p(train[s])
#check improved skewness
train[['VOTES','COST']].skew()
```

```python
#seperate target variable 
x = train.drop('COST',axis=1)
y = train['COST']
```

```python
#one hot encode cat data
x = pd.get_dummies(x)
x.head()
```

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#importing all the regression algorithms
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
#importing all the metrics and tuning methods for regression analysis
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=None)
#selecting the algorithm which works best 
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor(),
        RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),LGBMRegressor(silent=True),
        XGBRegressor(verbosity=0)]
for m in models:
    m.fit(x_train,y_train)
    print(m)
    print('SCORE :',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('r2_score : ',r2_score(y_test,pred))
    print('mean absolute error :',mean_absolute_error(y_test,pred))
    print('mean squared error :',mean_squared_error(y_test,pred))
    print('root mean squared error :',np.sqrt(mean_squared_error(y_test,pred)))
    print('\n')
```

After analysing above results, we can say that LGBMRegressor is giving less errors than others.

So finally, let us select LGBMRegressor & XGBRegressor and find each ones' max r2 score with best random state.

# LGBMRegressor

```python
#let us find max r2_score with corresponding random state
lgbr = LGBMRegressor(silent=True)
max_score_lgbr=0
best_ran_state_lgbr=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    lgbr.fit(x_train,y_train)
    pred=lgbr.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_lgbr < score:
        max_score_lgbr = score
        best_ran_state_lgbr = n
print('max r2_score of LGBRegressor : ',max_score_lgbr)
print('best random state : ',best_ran_state_lgbr)
```

# XGBRegressor

```python
#let us find max r2_score with corresponding random state
xgbr = XGBRegressor(verbosity=0)
max_score_xgbr=0
best_ran_state_xgbr=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=n)
    xgbr.fit(x_train,y_train)
    pred=xgbr.predict(x_test)
    score=r2_score(y_test,pred)
    if max_score_xgbr < score:
        max_score_xgbr = score
        best_ran_state_xgbr = n
print('max r2_score of XGBRegressor : ',max_score_xgbr)
print('best random state : ',best_ran_state_xgbr)
```

We can see that LGBM is giving the max r2_score. So, let us select LGBM for final training.

```python
#let's tune parameters of LGBRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=45)
param = {
        'max_depth': [5,6,7,8,9,10],'learning_rate' : [0.01,0.05,0.1],'num_leaves': [50,100,150,200],
        'n_estimators': [50,100,150,200,250,300]
        }
GCV_lgbr= GridSearchCV(lgbr,param_grid=param,cv=10,scoring='r2')
GCV_lgbr.fit(x_train,y_train)
```

```python
#get best params for lgbr
GCV_lgbr.best_params_
```

```python
#final training and testing with best parameters
lgbr_best = LGBMRegressor(learning_rate= 0.05,max_depth=10,n_estimators=200,num_leaves=50,silent=True)
lgbr_best.fit(x_train,y_train)
preds=lgbr_best.predict(x_test)
print('SCORE :',lgbr_best.score(x_train,y_train))
print('r2_score : ',r2_score(y_test,preds))
print('mean absolute error :',mean_absolute_error(y_test,preds))
print('mean squared error :',mean_squared_error(y_test,preds))
print('root mean squared error :',np.sqrt(mean_squared_error(y_test,preds)))
```

```python
#cross_validation
scores=cross_val_score(lgbr_best,x,y,cv=10,scoring='r2')
print('mean r2 score : ',scores.mean())
```

```python
#plotting predicted v/s actual values
sns.distplot(preds,label='predicted values')
sns.distplot(y_test,label='actual values')
plt.title('predicted v/s actual values')
plt.legend()
plt.show()
```

```python
#finally fiiting 100% data for deployment
lgbr_best.fit(x,y)
lgbr_best.score(x,y)
```
